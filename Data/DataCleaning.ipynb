```python
import pandas as pd
import re
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
#Remove invalid chars 
data = pd.read_csv('spotify.csv')

def remove_invalid_chars(s):
    return re.sub(r'[^\x20-\x7E]', '', str(s))

cleaned_data = data.applymap(remove_invalid_chars)

cleaned_data.to_csv('spotify_cleaned.csv', index=False)
```

```python
cleaned_data = pd.read_csv('spotify_cleaned.csv')
cleaned_data = cleaned_data.drop('Unnamed: 0', axis=1)
cleaned_data.to_csv('spotify_cleaned.csv', index=False)
```

```python
cleaned_data = pd.read_csv('spotify_cleaned.csv')
cleaned_data.drop_duplicates(subset='track_id', keep='first', inplace=True)
cleaned_data.drop_duplicates(subset=['track_name', 'artists'], keep='first', inplace=True)

cleaned_data.to_csv('spotify_cleaned.csv', index=False)
cleaned_data
```

```python
spotify = pd.read_csv('spotify_cleaned.csv')
spotify.drop(['track_name', 'artists', 'album_name'], axis=1, inplace=True)
spotify.head()
```

```python
numerical_columns = spotify.columns.drop(['popularity', 'track_id', 'track_genre'])
scaler = StandardScaler()
spotify[numerical_columns] = scaler.fit_transform(spotify[numerical_columns])
```

```python
encoder = OneHotEncoder(sparse_output=False) 
track_genres = spotify[['track_genre']]  
encoded_genres = encoder.fit_transform(track_genres)

feature_names = encoder.get_feature_names_out(['track_genre'])
encoded_genres_df = pd.DataFrame(encoded_genres, columns=feature_names)

encoded_genres_df.index = spotify.index

spotify_encoded = pd.concat([spotify.drop('track_genre', axis=1), encoded_genres_df], axis=1)
spotify_encoded.to_csv('spotify_encoded.csv', index=False)
spotify_encoded
```
